1. Дослідимо, який філіал супермаркету ('Branch') є найприбутковішим. Для цього знайдіть сумарний прибуток за кожним філіалом і виявіть, який філіал має найвищий.

```python
income_by_branches = df.groupby('Branch')['gross income'].sum()
income_by_branches
```

```python
income_by_branches.sort_values(ascending=False)[:1]
```

Відповідь: філіал С - найприбутковіший.

```python
# знаходимо індекс максимального значення (назву філіалу) для фільтрації
max_income_branch = income_by_branches.idxmax()
max_income_branch
```

**Питання №1**

Визначити найбільш та найменш популярні категорії продукції сумарно по всіх магазинах та середню вартість придбаних товарів, без урахування податку по кожній категорії.

```python
#Розраховуємо загальну кількість відвідувань по всім категоріям по всіх магазинах
most_popular_product_line = df.groupby('Product line').agg({'Invoice ID':'count'})

#Розраховуємо середню вартість придбаних товарів по кожній категорії
cogs_mean_rer_product_line = df.groupby('Product line').cogs.mean()
display(most_popular_product_line)

with pd.option_context("display.precision", 2):
    display(cogs_mean_rer_product_line)
```

```python
# Таблиця з кількістю інвойсів по категоріям товарів для жінок і чоловіків
invoices_by_product_line_and_gender = df.pivot_table(index=['Gender'], columns=['Product line'], values=['Invoice ID'], aggfunc='count')
display(invoices_by_product_line_and_gender)

# Відображаємо різницю між кількістю інвойсів в сусідніх категоріях товарів
display(invoices_by_product_line_and_gender.diff(1))
```

```python
def get_hour(time):
    return int(time.split(':')[0])
```

```python
df['Hour'] = df.Time.apply(get_hour)
```

```python
# Конвертуємо колонку 'Date' у формат datetime
df['Date'] = pd.to_datetime(df['Date'])

# Додаємо нову колонку 'Day_of_Week', яка містить день тижня
df['Day_of_Week'] = df['Date'].dt.day_name()

# Групування даних за днем тижня та часом
sales_by_day_time = df.groupby(['Day_of_Week', 'Hour'])['gross income'].sum()
```

```python
display(sales_by_day_time)
```

```python
# Знаходимо годину з максимальним прибутком для кожного дня тижня
max_sales_by_day_time = sales_by_day_time.groupby(level='Day_of_Week').idxmax()

print(max_sales_by_day_time)
```
