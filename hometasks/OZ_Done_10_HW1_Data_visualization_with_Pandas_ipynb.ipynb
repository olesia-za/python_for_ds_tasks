## Завдання 7: Порівняння користувачів (Stacked Bar Chart)

**Завдання:**
Ми хочемо дізнатись як по дням тижня беруть в середньому в оренду велосипеди випадкові і зареєстровані користувачі.

Створіть стовпчасту діаграму з накопиченням (bar з налаштуванням `stacked=True`), яка показує співвідношення випадкових (`casual`) та зареєстрованих (`registered`) користувачів по днях тижня за кількістю взятих ними велосипедів в оренду в середньому. Використайте різні кольори для типів користувачів.

**Питання для інтерпретації:**
1. В які дні тижня більше оренд від зареєстрованих користувачів?
2. Як ви можете пояснити таку різницю в поведінці користувачів протягом тижня?

```python
ax = df.groupby('weekday')[['casual','registered']].agg(['mean']).plot.bar(
    stacked=True,
    figsize=(12, 6),
    title='Середня кількість оренди велосипедів по дням тижня',
    xlabel='День тижня',
    ylabel='Cередня кількість оренди',
)
ax.legend (title ='Тип користувача', labels=['Прокат для випадкових користувачів', 'Прокат для зареєстрованих користувачів'], loc ='center right', bbox_to_anchor=(1.2, 1.1))
plt.show();
# 1) більше орендують зареєстровані користувачі в усі дні тижня, але у вихідні дні (субота та неділя) - середня кількість оренди від випадкових користувачів збільшується
# 2) у вихідні дні середня кількість оренди від випадкових користувачів збільшується, що може свідчити про те, що в будні дні більше сервісом користуються місцеві жителі 
# або ті хто регулярно користується сервісом,а у вихідні - туристи або люди, які не користуються сервісом регулярно
```
